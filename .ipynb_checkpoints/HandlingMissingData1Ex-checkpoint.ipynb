```python
import pandas as pd
```

```python
df = pd.read_csv('fruits_data.csv',parse_dates = ['date'])
df
```

```python
df.shape
```

```python
df.columns
```

```python
new_df = df.fillna(-1)
new_df
```

```python
#write the code to fill null values as per above condition
#store the result to 'new_df' variable
new_df = df.fillna({
        'apple(1kg)': df['apple(1kg)'].mean(),
        'banana(1 dozen)': df['banana(1 dozen)'].mean(),
        'grapes(1kg)': df['grapes(1kg)'].median(),
        'mango(1kg)': df['mango(1kg)'].median(),
        'Water Melons(1)': "Not Available",
        
    })

#show the dataframe 'new_df'
new_df
```

```python
new_df = df.fillna(method = 'ffill')
new_df
```

```python
#write the code here
new_df = df.dropna(thresh=4)

#show the dataframe 'new_df'
new_df
```

```python
#write the code here
new_df = df.dropna()


#show the dataframe 'new_df'
new_df
```

```python
#save the dataframe to 'final_data.csv'
new_df.to_csv('final_data.csv', index = False)
```
