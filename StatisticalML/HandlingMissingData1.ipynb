```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv",parse_dates = ['day'])
df
```

```python
type(df.temperature[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature' : df['temperature'].mean(),
    'windspeed' : df.windspeed.mean(),
    'event': 'No Event'
})
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df

#how='all': Drop the row only if all columns are NaN

#If even one column has a non-null value, the row is kept
```

```python
new_df = df.dropna(how = 'any')
new_df
```

```python
new_df = df.dropna(thresh=2) # When a row has two columns with NaN values
new_df
```
